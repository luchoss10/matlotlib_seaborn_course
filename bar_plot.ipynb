```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population,width=0.5,color=['red','blue'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(8,8))
axs.bar(country,population,width=0.5,color=['orange','blue','green','violet','red'])
axs.set_title("País - Población")
axs.set_xlabel("País")
axs.set_ylabel("Población")
axs.set_xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)
fig.show()
```
